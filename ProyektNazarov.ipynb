## II. Датасет
https://www.kaggle.com/datasets/rajugc/imdb-top-250-movies-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/IMDB Top 250 Movies.csv')
df.head()
```

```python
df.info()
```

```python
#Популярные жанры
popular_genres = df['genre'].value_counts().head(10)

#Рейтинги фильмов
average_ratings = df.groupby('genre')['rating'].mean()

#Кассовые сборы
box_office = df.groupby('genre')['box_office'].sum()
```

```python
#Визуализация популярных жанров
popular_genres.plot(kind='bar', title='Top 10 Popular Genres')
plt.xlabel('genre')
plt.ylabel('Number of Movies')
plt.show()

#Визуализация средних рейтингов по жанрам
average_ratings.plot(kind='bar', title='Average Ratings by Genre')
plt.xlabel('genre')
plt.ylabel('rating')
plt.show()

#Визуализация кассовых сборов по жанрам
box_office.plot(kind='bar', title='Box Office by Genre')
plt.xlabel('genre')
plt.ylabel('box_office')
plt.show()
```

```python
print(df.columns)
```

```python
# Показать самые кассовые фильмы
top_grossing_movies = df[['name', 'box_office']].sort_values('box_office', ascending=False).head(10)
print("Самые кассовые фильмы:")
print(top_grossing_movies)
```

```python
# Показать самый популярный жанр
most_popular_genre = df['genre'].value_counts().idxmax()
print("Самый популярный жанр:", most_popular_genre)
```

```python
# Показать самые высокие по рейтингу фильмы
top_rated_movies = df[['name', 'rating']].sort_values('rating', ascending=False).head(10)
print("Самые высокие по рейтингу фильмы:")
print(top_rated_movies)
```

```python
import matplotlib.pyplot as plt

# Самые кассовые фильмы
plt.figure(figsize=(12, 6))
plt.bar(top_grossing_movies['name'], top_grossing_movies['box_office'], color='blue')
plt.title('Top Grossing Movies')
plt.xlabel('Movie Name')
plt.ylabel('Box Office ($)')
plt.xticks(rotation=90)
plt.show()

# Распределение по жанрам
plt.figure(figsize=(12, 6))
df['genre'].value_counts().plot(kind='bar', color='green')
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()

# Самые высокие по рейтингу фильмы
plt.figure(figsize=(12, 6))
plt.bar(top_rated_movies['name'], top_rated_movies['rating'], color='orange')
plt.title('Top Rated Movies')
plt.xlabel('Movie Name')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Самые провальные фильмы по кассовым сборам
worst_box_office_movies = df.sort_values('box_office')[:10]

#Самые провальные фильмы по рейтингу
worst_rated_movies = df.sort_values('rating')[:10]

plt.figure(figsize=(12, 6))

#Cамые провальные фильмы по кассовым сборам
plt.subplot(1, 2, 1)
plt.bar(worst_box_office_movies['name'], worst_box_office_movies['box_office'], color='red')
plt.title('Worst Box Office Movies')
plt.xlabel('Movie Name')
plt.ylabel('Box Office ($)')
plt.xticks(rotation=45, ha='right')

#Самые провальные фильмы по рейтингу
plt.subplot(1, 2, 2)
plt.bar(worst_rated_movies['name'], worst_rated_movies['rating'], color='purple')
plt.title('Worst Rated Movies')
plt.xlabel('Movie Name')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/IMDB Top 250 Movies.csv')
df['year'] = pd.to_datetime(df['year'], errors='coerce')
genre_year = df.groupby(['year', 'genre']).size().unstack().fillna(0)
genre_year_norm = genre_year.div(genre_year.sum(axis=1), axis=0)

plt.figure(figsize=(12, 8))
sns.set_theme(style="darkgrid")
ax = sns.lineplot(data=genre_year_norm[::10], dashes=False)

plt.title('Динамика популярности жанров с 1921 по 2022 год')
plt.xlabel('Год')
plt.ylabel('Доля фильмов по жанрам')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```
